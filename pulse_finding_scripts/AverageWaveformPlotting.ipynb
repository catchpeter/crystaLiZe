```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tscale = (8.0/4096.0)

data_dirs = []
wfms = []
labels = []

data_dirs.append('G:/My Drive/crystalize/data/data-202104/041421/Po_0g_0c_1.1bar_circ_20min_1941/')
wfms.append(np.loadtxt(data_dirs[-1]+'average_waveform.txt'))
labels.append('LXe')

data_dirs.append('G:/My Drive/crystalize/data/data-202104/041921/Po_0g_0c_0.74bar_circ_20min_2131/')
wfms.append(np.loadtxt(data_dirs[-1]+'average_waveform.txt'))
labels.append('SXe')

times = np.array(range(len(wfms[0])))*tscale
```

```python
t_start = 6100 #samples
t_end = 6900
t = times[t_start:t_end]-times[t_start]

plt.figure()
for ii, wfm in enumerate(wfms):
    plt.plot(t, wfm[t_start:t_end], label=labels[ii])
    
plt.xlabel('Time (us)')
plt.legend()

# normalized version
plt.figure()
for ii, wfm in enumerate(wfms):
    wfm_norm = wfm/np.sum(wfm)
    plt.plot(t, wfm_norm[t_start:t_end], label=labels[ii]+' (unit area)')
    
plt.xlabel('Time (us)')
plt.legend()
```
